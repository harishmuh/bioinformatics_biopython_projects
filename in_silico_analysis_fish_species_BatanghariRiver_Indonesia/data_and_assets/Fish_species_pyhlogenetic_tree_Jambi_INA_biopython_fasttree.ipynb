# **In Silico Phylogenetic Analysis of Fish Species in the Batanghari River, Jambi, Indonesia using COI 1 Sequences**

## **Context**

### **Background**

The Batanghari River, located in central Sumatra, Indonesia, is one of the region’s largest and most ecologically significant rivers. Its basin spans multiple provinces and encompasses diverse habitats, including freshwater lakes, wetlands, and mangrove ecosystems. These varied habitats support a remarkable diversity of aquatic life, particularly fish species. However, growing human activities such as deforestation, urbanization, and pollution pose significant threats to the river's biodiversity.

Preserving the biodiversity of the Batanghari River is vital, not only for ecological stability but also for maintaining the river’s role in supporting local livelihoods and regional ecosystems. Additionally, understanding the relationships among fish species may offer critical insights into their ecological roles, which can guide effective conservation strategies.

**Challenges in Fish Species Identification**

Historically, fish species in the Batanghari River have been identified using morphological methods that rely on observable physical traits such as body shape, fin structures, scale patterns, and coloration. While effective for distinguishing many species, these methods are:

* Subjective, depending on the observer's expertise.
* Limited in distinguishing closely related species with similar external traits.

**The Role of Molecular Phylogenetics**

To address these limitations, molecular phylogenetics has become a valuable approach. By analyzing DNA sequences, researchers can infer genetic relationships between species, even when they appear morphologically similar.

This study focuses on the **cytochrome c oxidase subunit I (COI) gene**, a mitochondrial DNA marker widely used in phylogenetic studies of animals. The COI gene is ideal for this purpose because:

* It is maternally inherited.
* It has a relatively high mutation rate, making it effective for distinguishing closely related species.
* It is abundant in cells, ensuring high-quality data for analysis.

**In Silico - Study Basis**

The foundation for this phylogenetic analysis is the study by Nurdawati (2007), which explored the diversity and distribution of fish species in the Batanghari River. Sampling sites included various lakes and rivers in Jambi, such as Teluk Lake, Buluran Kenali Lake, Kaos Lake, and the Terap, Lubuk Ruso, and Pijoan Rivers. Nurdawati identified:

* 90 fish species, spanning 44 genera, 18 families, and 6 orders.
  
For this silico phylogenetic analysis analysis, 20 freshwater fish species were selected from Nurdawati’s findings. Their genetic data will be retrieved from the NCBI GenBank database. Although the sequences are not from specimens directly collected in the Batanghari River, they are expected to represent genetically similar populations, providing valuable insights into the evolutionary relationships among these species. This in silico approach offers a cost-effective and time-efficient alternative to traditional wet lab experiments. By leveraging bioinformatics tools, the study can analyze fish species' evolutionary relationships without requiring DNA extraction or sequencing.

### **Objective of the Analysis**
The main objectives of this in silico study are:

* To construct a phylogenetic tree using Biopython and FastTree libraries.
* To determine the evolutionary relationships among fish species from various habitats in the Batanghari River basin.

**Data Source**
* Scientific Basis: Nurdawati, S. (2007). Keanekaragaman dan Distribusi Benih Ikan di Beberapa Tipe Habitat Sungai Batanghari, Jambi. J. Lit. Perikan. Ind. Vol.13.
* Genetic Data: DNA sequences for the selected fish species will be obtained from the NCBI GenBank database.

```python
# Revision
# Load sequences from your original FASTA file
input_file = "sequences.fasta"
output_file = "sequences_with_species.fasta"

# Modify the sequence IDs and save to a new file
with open(output_file, "w") as out_fasta:
    for record in SeqIO.parse(input_file, "fasta"):
        # Replace the sequence ID with species name if found in the dictionary
        if record.id in id_to_species:
            id_number = record.id 
            full_species_name = id_to_species[record.id]
            record.id = full_species_name          # For use as the primary ID
            record.name = full_species_name        # For the name attribute
            record.description = ""                # Clear any additional description
        SeqIO.write(record, out_fasta, "fasta")
```

```python
# Create a function to read the metadata of a FASTA file
# Input is the filepath(directory address)
def sequence_file_info(file_path):
    sequences = SeqIO.parse(file_path, "fasta")
    
    for record in sequences:
        seq_id = record.id
        seq_name = record.name
        seq_description = record.description
        seq_record = record.seq
        seq_length = len(seq_record)
        
        # Display sequence information
        print(f"ID: {id_number}")
        print(f"Species Name: {seq_description}")
        print(f"Sequence Length: {seq_length}")
        print(' ')
```

### **Sequence Alignment**

The collected sequences were aligned using ClustalW (implemented within Biopython library) to identify homologous positions and prepare them for phylogenetic analysis.

```python
# Step 1: Perform alignment using ClustalW2
subprocess.run(["clustalw2", "-INFILE=sequences_with_species.fasta", "-ALIGN", "-OUTFILE=sequences.aln"])

# Step 2: Load the CLUSTAL alignment # Produce 'file format.aln'
alignment = AlignIO.read("sequences.aln", "clustal")
```

```python
# Step 3: Write alignment and convert the alignment to FASTA format
AlignIO.write(alignment, "converted_sequences.fasta", "fasta")
alignment_file = r"converted_sequences.fasta" # The converted FASTA alignment file
```

### **Phylogenetic tree construction**

**Phylogenetic Analysis**

For constructing phylogenetic tree we set several parameters of analysis that consist of:

* **Evolutionary Model Selection**
  
  The Kimura 2-parameter model was chosen to account for transitions and transversions in nucleotide substitutions.
* **Maximum Likelihood**
  
  The maximum likelihood method was employed to estimate the most likely evolutionary history of the fish species.
* **Bootstrap Analysis**
  
  Bootstrap analysis with 100 replicates was performed to assess the reliability for the inferred relationships.


**Tree Construction**

* The FastTree library was used to construct the phylogenetic tree based on the maximum likelihood results.

```python
# Phlyogenetic tree construction using The Kimura 2-parameter model, bootstrap analysis
result = subprocess.run(
    [r'C:\Program Files (x86)\FastTree\FastTree.exe', '-nt', '-gtr', '-boot', '100', alignment_file],
    capture_output=True, text=True
)
```

```python
# Checking if FastTree is working correctly
if result.returncode == 0:
    print("FastTree is working!")
    print("Bootstrap and tree generation complete!")
    print('')
    print(f'Output tree: {result.stdout}')  # Output tree in Newick format
else:
    print("FastTree is not working correctly.")
    print(result.stderr)  # Print error message if there's an issue
```
